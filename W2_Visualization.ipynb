```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1au_UL4MpOzO9ln_6cuIC6Oa_Kozuel3nBtIadjy5Xr4/edit?usp=sharing')
```

```python
in_stock = {
    "Food": {
        "Rice": 10,
        "Bread": 6

    },
    "Supplies": {
        "Nails": 54,
        "Screws": 61

    }
}
```

```python
in_stock['Food'].keys()
```

```python
import pandas as pd

df2=pd.DataFrame(list(in_stock.keys()),columns=['Item'])
```

```python
df2
```

```python
data=[]
for category, items in in_stock.items():
    for item, quantity in items.items():
        data.append([category, item, quantity])

df_all=pd.DataFrame(data,columns=['Category', 'Item', 'Stock'])
```

```python
df_all
```

```python
from gspread_dataframe import set_with_dataframe

# 選擇要更新的工作表（選擇第一個工作表）
worksheet = gsheets.get_worksheet(0)  # 如果是其他工作表，可以用 get_worksheet(index)

# 將 DataFrame 寫入 Google Sheets
set_with_dataframe(worksheet, df_all)
```

```python
worksheet_scores=gsheets.get_worksheet(1)
```

```python
worksheet_scores
```

```python
# 取得 Google Sheets 內容
data = worksheet_scores.get_all_values()  # 取得所有儲存格資料，回傳為 2D 陣列

# 將資料轉換為 DataFrame
df_scores = pd.DataFrame(data)
```

```python
df_scores.columns=df_scores.iloc[0]
df_scores=df_scores.iloc[1:]

df_scores.reset_index(drop=True, inplace=True)
```

```python
df_scores
```

```python
df_ENscores=df_scores.drop(columns='數學')
df_ENscores
```
