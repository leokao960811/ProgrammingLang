所有產品價位分布

```python
import plotly.express as px
import plotly.io as pio
import numpy as np

# Define bins and labels
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
labels = ["0-50", "51-100", "101-150", "151-200", "201-250", "251-300", "301-350", "351-400", "401-451", "451-500"]

# Create a new column with score ranges
df_adjusted['Price Range'] = pd.cut(df_adjusted['Price'], bins=bins, labels=labels, right=True)

# Count students in each range
score_counts = df_adjusted['Price Range'].value_counts().sort_index()

# Plot the bar chart
fig = px.bar(x=score_counts.index, y=score_counts.values,
             labels={'x': '價錢範圍', 'y': '數量'},
             title="產品價位分布",
             color=score_counts.values, color_continuous_scale="blues")

fig.show()
html_graph1 = pio.to_html(fig)
```

##圖表現象與趨勢：

#### 1.  產品分布均勻

大部分價位區間內的產品數量相當接近，約在 1000 件左右，表示產品的價格分佈較平均，沒有明顯集中。

#### 2.  價格 0-50 和 301-350 的區間，產品數量較低

0-50 區間的產品數量明顯低於 1000，顯示低價產品相對較少。而301-350 區間的產品數量也低於其他區間，可能代表這個價格帶的產品較少，或者市場需求較低。

#### 3.  高價位（400-500）產品數量最大。


##可能的商業意義
#### 1.  低價位產品的市場空間

由於 0-50 價位的產品數量較少，可能代表市場上較少廠商提供超低價產品，可能是一個未被充分開發的機會。

#### 2.  301-350 區間的特殊性

這個區間的產品數量較少，有可能是因為該價位帶的需求較低，或者消費者更傾向選擇較便宜或較昂貴的產品，可進一步調查。

#### 3.  高價位市場仍有競爭力

451-500 價位的產品仍有相當數量，表示高價市場仍然有足夠的供應與需求，可能適合高品質或品牌定位的產品。

不同價位產品的5星評價數

```python
# Assuming the dataframe is already loaded into df and contains columns like 'actor.name', 'Date', 'verb.display.zh-TW', 'theme'

# Filter out rows where theme is None (if applicable)
df_filtered = df_adjusted[df_adjusted['Product Category']=='Home Appliances']

# Group by Product Name','Price Range', and 'Product Ratings'
product_count = df_filtered.groupby(['Product Name','Price Range','Product Ratings']).size().unstack(fill_value=0)

# Display the counts per rating
product_count
```

```python
names = product_count.index.get_level_values('Product Name')
range = product_count.index.get_level_values('Price Range')

# Example: Combine into a DataFrame for easy viewing
df_with_index = pd.DataFrame({
    'Product Name': names,
    'Price Range': range,
})
```

```python
df_with_index
```

```python
merged_df = pd.merge(df_with_index, product_count, on=['Product Name', 'Price Range'])
```

```python
merged_df
```

```python
import plotly.express as px

# Group the data by Product Name and Price Range, summing the 5-star counts
df_rate_counts = merged_df.groupby(['Product Name', 'Price Range'])[5].sum().reset_index()

# Plot using Plotly
fig = px.bar(df_rate_counts, x='Product Name', y=5, color='Price Range',
             title="5-Star Ratings Across Different Product Types",
             labels={'5': '5-Star Ratings', 'Price Range': 'Price Range'},
             barmode='stack')

# Update layout for better visualization
fig.update_layout(xaxis_title="Product Types", yaxis_title="5-Star Ratings",
                  legend_title="Theme", title_x=0.5)

# Show the plot
fig.show()
html_graph2 = pio.to_html(fig)
```

```python
prompt=f"以下為一個圖表的HTML：{html_graph2}，請分析此圖表，觀察現象及趨勢。"
response = model.generate_content(prompt)
print(response.text)
```

```python
import plotly.express as px
import plotly.graph_objects as go

# 使用數據透視表進行重新排列表格
pivot_subject_behavior = merged_df.pivot_table(index='Product Name', columns='Price Range', values=5, fill_value=0)

# 使用 Plotly 的熱力圖繪製
fig = go.Figure(data=go.Heatmap(
    z=pivot_subject_behavior.values,
    x=pivot_subject_behavior.columns,
    y=pivot_subject_behavior.index,
    colorscale='YlGnBu'
))

# 設定圖表標題和軸標籤
fig.update_layout(
    title="所有產品在不同價位下的5星評論數",
    xaxis_title="價位",
    yaxis_title="產品",
    height=800,  # 設定圖表高度
    width=1000   # 設定圖表寬度
)

# 顯示圖表
fig.show()
html_graph3 = pio.to_html(fig)
```

##圖表現象與趨勢：

### 1. 高評價的集中區域

Monitor 在 51-100 有最多 5 星評論；
Laptop（筆電）在 301-350 和 451-500 價位區間獲得較多 5 星評論；
Smartphone（智慧型手機）在 401-451 及 451-500 價位的 5 星評論數較多。

可能表示這些產品在這些價格範圍內有較高的性能、品質，受到消費者的青睞。

### 2. 低評價的集中區域
Monitor 在 251-300 價位的 5 星評論數最低（幾乎沒有）。

可能這個價位的顯示器產品評價較不佳，或者市場上的選擇較少，導致高評價數量偏低。

Laptop 在 151-200 價位區間的 5 星評論數較低。

可能是因為這個價位的筆電規格較入門，滿足不了大多數使用者的需求，導致評價相對較低。

Headphones（耳機）整體分布較均勻，但 0-50 價位區間的 5 星評論數偏少。

低價耳機可能因品質或音質較差，較難獲得高評價。

## 可能的商業意義
####高價產品更容易獲得 5 星評價

智慧型手機、筆電和部分耳機在較高價位區間（400 以上）獲得較多的 5 星評論，顯示消費者更願意為高價產品給予正面評價。

####某些價位區間的產品評價明顯較差

例如 251-300 的顯示器和 151-200 的筆電，這可能代表這些價位的產品需要改進，或者消費者不滿意該價位提供的選擇。

####高性價比產品可能集中在中等價位

顯示器在 51-100 區間的 5 星評價特別高，這可能代表這個價位的產品最具性價比，值得企業專注於這個市場區間。

KMeans 分類、分析

```python
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 計算每個學生的行為數據
# student_behavior_summary = df_adjusted.groupby('Product Name').agg({'Stock Quantity': 'sum', 'Product Ratings': 'sum'}).reset_index()

# 使用KMeans進行聚類
kmeans = KMeans(n_clusters=4)
merged_df['cluster'] = kmeans.fit_predict(merged_df[[1, 5]])

# 使用PCA進行降維視覺化
pca = PCA(n_components=2)
components = pca.fit_transform(merged_df[[1, 5]])
merged_df['pca1'] = components[:, 0]
merged_df['pca2'] = components[:, 1]

# 視覺化學生行為聚類結果
fig = px.scatter(
    merged_df,
    x='pca1',
    y='pca2',
    color='cluster',
    title="產品聚類結果",
    labels={'pca1': '主成分1', 'pca2': '主成分2', 'cluster': '聚類群組'},
    hover_data=['Product Name', 1, 5]  # 在互動中顯示更多細節
)

# 更新圖表的佈局
fig.update_layout(
    xaxis_title="主成分1",
    yaxis_title="主成分2",
    legend_title="群組",
    title_x=0.5,  # 圖表標題居中
    height=600,
    width=800
)

# 顯示圖表
fig.show()
html_graph4 = pio.to_html(fig)
```
